### 1.7: Who is the most common type of room, in each neighbourhood?

```python
Finaldf
Final=Finaldf.groupby(['id','neighbourhood','room_type'])
Final=Final.sum()
Final=Final.reset_index()
Final

Current=Final

Neighbourhoods=Final['neighbourhood'].unique()
Neighbourhoods


Dataset=pd.DataFrame(data=Neighbourhoods, columns=['Neighbourhood'])

Dataset=Dataset.assign(Entire_room_apt=0)
Dataset=Dataset.assign(Private_room=0)
Dataset=Dataset.assign(Shared_room=0)
Dataset=Dataset.set_index('Neighbourhood')


# Dataset=Dataset.rename(columns={"Entire_room_apt":"Entire room/apt" , "Private_room": "Private room", "Shared_room":"Shared room"})
Dataset

#We'll fill this dataset with the right values
```

```python
for n in range (len(Neighbourhoods)):
    print (Neighbourhoods[n])
    Current=Final[Final.neighbourhood.eq(Neighbourhoods[n])]
    Current=Current['room_type'].value_counts().rename_axis('room_type').reset_index(name="counter")
   
    for i in range (len(Current)):
#        print (Current)
#        print("\n")
        
        
        if Current["room_type"][i]=="Entire home/apt":
            Dataset['Entire_room_apt'][n]=Current['counter'][i]
        
        elif Current["room_type"][i]=="Private room":
            Dataset['Private_room'][n]=Current['counter'][i]
        
        elif Current["room_type"][i]=="Shared room":
            Dataset['Shared_room'][n]=Current['counter'][i]

#Non-graphic content:
Dataset

```

```python
PlotEntire=Dataset[['Entire_room_apt']]
PlotEntire=PlotEntire.sort_values(by='Entire_room_apt',ascending=False)
PlotEntire.plot.bar()

print('            1)Plot for Entire rooms')


PlotPrivate=Dataset[['Private_room']]
PlotPrivate=PlotPrivate.sort_values(by='Private_room',ascending=False)
print('            2)Plot for Private rooms')
PlotPrivate.plot.bar()

PlotShared=Dataset[['Shared_room']]
PlotShared=PlotShared.sort_values(by='Shared_room',ascending=False)
print('            2)Plot for Shared rooms')
PlotShared.plot.bar()
```
